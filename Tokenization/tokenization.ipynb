```python
## Word Tokenization
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt_tab')
```

```python
text = "i like to eat apples and bananas"
tokens = word_tokenize(text)
```

```python
print(tokens)
```

```python
## Sentence Tokenization
```

```python
from nltk.tokenize import sent_tokenize
text_1 = "im gonna tokenize this string. for a machine learning model, right? "
```

```python
sentences = sent_tokenize(text_1)
```

```python
for i, sentence in enumerate(sentences, 1):
    print(f"Sentence {i}: {sentence}")
```

```python
## Token frequency counter
```

```python
from collections import Counter
```

```python
text_2 = "My code works. I have no idea why my code is running"
```

```python
tokens = word_tokenize(text_2.lower())
frequency = Counter(tokens)
```

```python
print(frequency)
```
